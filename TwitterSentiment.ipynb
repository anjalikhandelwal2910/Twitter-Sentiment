```python
import string
punc = list(string.punctuation)
```

```python
from nltk.corpus import wordnet
from nltk import pos_tag
```

```python
from nltk.stem import WordNetLemmatizer
lemm = WordNetLemmatizer()
```

```python
tknzr = TweetTokenizer()
```

```python
xtrain = []
for i in range(len(x_train)):
    w = tknzr.tokenize(x_train[i][0])
    xtrain.append(w)
```

```python
x_train = []
for i in range(len(xtrain)):
    x_train.append((xtrain[i], y[i][0]))
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
stop_words.update(punc)
#stop_words.update(air)
#stop_words.update("http://t.co/413gial0yl")
```

```python
def get_simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
def clean_tweet(words):
    output_words=[]
    for w in words:
        #if not w.startswith("http:"):
        if w.lower() not in stop_words:
            pos = pos_tag([w])
            clean_word = lemm.lemmatize(w, pos= get_simple_pos(pos[0][1]))
            output_words.append(clean_word.lower())
    return output_words
```

```python
train_data = [(clean_tweet(doc), category) for doc, category in x_train]
```

```python
train_data
```

```python
all_words =[]
for doc in train_data:
    all_words+=doc[0]
```

```python
import nltk
```

```python
freq = nltk.FreqDist(all_words)
common = freq.most_common(7800)
features = [i[0] for i in common]
```

```python
def get_feature_dic(words):
    current_features ={}
    words_set = set(words)
    for w in features:
        current_features[w] = w in words_set
    return current_features
```

```python
train_data = [(get_feature_dic(doc), category) for doc , category in train_data]
```

```python
len(train_data)
```

```python
from nltk import NaiveBayesClassifier
```

```python
clf = NaiveBayesClassifier.train(train_data)
```
